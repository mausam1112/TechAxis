## Data Distribution

```python
import numpy as np
import seaborn as sns
```

```python
# Uniform distribution
def create_dataframe(n, dist=''):
    if dist == 'uniform':
        data = np.random.uniform(0, 1, n)
    elif dist == 'normal':
        data = np.random.normal(0, 1, n)
    elif dist == 'exp':
        data = np.random.exponential(1, n)
    elif dist == 'log':
        data = np.random.lognormal(0, 1, n)
    df = pd.DataFrame({
        'index':  np.arange(1000),
        'data': data
    })
    return df
```

```python
df = create_dataframe(1000, dist='uniform')
sns.histplot(df['data'])
```

```python
import matplotlib.pyplot as plt
df = create_dataframe(1000, dist='normal')
sns.histplot(df['data'], kde=True)
```

```python
df = create_dataframe(1000, dist='exp')
sns.histplot(df['data'], kde=True)
```

```python
df = create_dataframe(1000, dist='log')
sns.histplot(df['data'], kde=True)
```

```python
df = create_dataframe(1000, dist='exp')
sns.histplot(df['data'] * (-1), kde=True)
```
