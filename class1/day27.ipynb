# Subplots	&	Multiple	Plots	in	Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 100, 10)
x
```

```python
np.sin(x), np.cos(x)
```

# .subplot()

```python
plt.figure(figsize=(6, 3))
plt.plot(x)
plt.show()
```

```python
plt.plot(np.sin(x))
```

```python
plt.subplot(2, 2, 1)
plt.scatter(np.arange(len(x)), x)
plt.xlabel('n_x')
plt.ylabel('x')
plt.title('n_x vs X')


plt.subplot(2, 2, 2)
plt.plot(x)
plt.ylabel('x')
plt.title("Line plot")

plt.subplot(2, 2, 3)

plt.subplot(2, 2, 4)
plt.scatter(
    np.arange(len(x)), 
    np.exp(x)
)
plt.xlabel('n_x')
plt.ylabel('exp(x)')
plt.title('n_x vs exp(x)')

plt.suptitle("Age")

plt.tight_layout()
plt.show()
```

### .subplots()

```python
fig, ax = plt.subplots(
    3, 2, sharex=True, sharey=True
)

ax[0, 0].scatter(np.arange(len(x)), np.sin(x))
ax[0, 0].set_title('Scatter plot')
ax[0, 1].scatter(np.arange(len(x)), np.cos(x))
ax[0, 1].set_title('Scatter plot')

ax[1, 0].scatter(np.arange(len(x)), np.cos(x))
ax[1, 0].set_title('Scatter plot')
ax[1, 1].scatter(np.arange(len(x)), np.cos(x))
ax[1, 1].set_title('Scatter plot')

ax[2, 0].plot(np.sin(x))
ax[2, 0].set_title('Sine plot')
ax[2, 1].plot(np.sin(x))
ax[2, 1].set_title('Sine plot')


plt.tight_layout()
fig.suptitle('jdfasdfa')

fig.savefig('temp.png', transparent=False)
```
