## Nominal Data Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue', 'Black', 'Blue']
})
df
```

- One-Hot Encoding for Nominal Data

```python
color_encoded = pd.get_dummies(df, columns=['Color'], drop_first=True).astype(int)
color_encoded
```

- Label Encoding for Ordinal Data

```python
df = pd.DataFrame({
    "Grade": ['A', 'A+', 'B', 'C', 'B', 'A', 'B+']
})
df
```

```python
mapping = {'C':0, 'B':1, 'B+':2, 'A':3, 'A+':4}
```

```python
df['Grade_en'] = df['Grade'].map(mapping)
df
```

```python
from sklearn.preprocessing import LabelEncoder, OrdinalEncoder
```

```python
le = LabelEncoder()

df['Grade_en_le'] = le.fit_transform(df['Grade'])
df
```

```python
grade_hierarchy = [['C', 'B', 'B+', 'A', 'A+']]
ord_en = OrdinalEncoder(categories=grade_hierarchy)

df[['Grade_en_ord']] = ord_en.fit_transform(df[['Grade']])
df
```
