### Correlation
- ranges from -1 to +1
- when value is -1 (-ve correlation), if one variable increases, the other decreases.
- when value is +1 (+ve correlation), if one variable increases, the other also increases.
- when value is 0 (neutral/ no correlation), if one variable changes, other varibale remains constant (no changes).

```python
corr_mat = df.drop('species', axis=1).corr()
corr_mat
```

```python
sns.heatmap(corr_mat, annot=True, cmap='coolwarm')
```
