```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
arr = np.random.randint(0, 100, 200)
arr
```

```python
arr2 =  np.random.randint(0, 100, 200)
```

```python
plt.scatter(
    np.arange(200), 
    arr, 
    s=10, 
    c='red', 
    marker='D',
    label='male'
)
plt.scatter(np.arange(200), arr2, s=10, c='green', marker='*', label='female')
plt.legend()
plt.xlabel('Earnings in thousands')
plt.ylabel('Age')
plt.title("Earnings wrt Age")
plt.grid(axis='both')
```

```python
arr = [2, 3, 4, 8, 1, 7]
plt.figure(figsize=(10, 4))
plt.plot(arr, marker='*', mec='red', mfc='black')
plt.xlabel('Number of people')
plt.ylabel('Income in millions')
plt.title("Earnings wrt Number of people")
plt.grid(axis='both')
```

```python
plt.plot(arr, '*', mec='red', mfc='black')
```
